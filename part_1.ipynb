### <span style="color:turquoise">3. Förklara vad nedanstående begrepp, kopplat till klasser, betyder: </span>
#### <span style="color:yellow">a) Instans -  </span> ***en kopia av klassen, ex p1 = Person("Emma").***
#### <span style="color:yellow">b) Attribut - </span> ***data som är kopplat till instansen, 'namn' är i detta fallet ett attribut.***
#### <span style="color:yellow">c) Metod - </span> ***en funktion som är definerad inne i en klass, t.ex säg_hej().***

```python
class Person:
    def __init__(self, namn):
        self.namn = namn ## Attribut

    def säg_hej(self): ## Metod
        print(f"Hej, jag heter {self.namn}")

p = Person("Emma") ## Instans
p.säg_hej()
```

### <span style="color:turquoise"> 5. Skapa en klass som heter BankAccount. Klassen skall ha attributet "account_holder" som visar kontoinnehavarens namn samt attributet "balance" som visar kontoinnehavarens balans. Klassen skall ha metoden "deposit()" för att kunna sätta in pengar på kontot samt metoden "withdraw()" för att kunna ta ut pengar från kontot. Om bankinnehavaren försöker ta ut mer pengar än vad som finns på kontot skall meddelandet "Too low balance" printas ut. </span>

```python
class BankAccount:
  def __init__(self, account_holder, balance=0):
    self.account_holder = account_holder
    self.balance = balance

  def deposit(self, amount):
    self.balance += amount

  def withdraw(self, amount):
    if amount > self.balance:
      print('Saldo otillräckligt')
    else: 
      self.balance -= amount

# Test av klassen
account = BankAccount("Kevin")
print(account.account_holder)
print('Saldo:', account.balance)

account.deposit(500)
print('Efter insättning:', account.balance)

account.withdraw(200)
print('After uttag:', account.balance)

account.withdraw(1000) # Otillräckliga medel
```

### <span style="color:turquoise">6.Skapa en instans av klassen och testa så klassen funkar så som du förväntar dig. Du kan till exempel prova printa ut attributen, sätta in pengar och ta ut pengar. På intervjuer brukar ibland kodtest genomföras. Här är ett exempel på en fråga som hade kunnat dyka upp på ett kodtest.  Skapa en Python funktion som beräknar antalet vokaler i en sträng. Vokalerna är: A, E, I, O, U, Y, Å, Ä och Ö. </span>


```python
def vowel_checker(word):
  vowels = "aeiouyåäöAEIOUYÅÄÖ"
  return sum(1 for letter in word if letter in vowels)

vowel_checker('hjÄrna') # returnerer 2
```

### <span style="color:turquoise">7. Exempelvis: vowel_checker(’hjärna’) ska returnera ”2” eftersom ordet ”hjärna” har de två vokalerna ”ä” och ”a”. På intervjuer brukar ibland kodtest genomföras. Här är ett exempel på en fråga som hade kunnat dyka upp på ett kodtest. Skapa en funktion som tar två listor som argument och returnerar de gemensamma elementen i listorna i en ny lista.  Exempelvis:  common_elements([4, ’apple’, 10, ’hi’, 3], ([23 ’apple’, 5, 9, 3]) ska returnera svaret [’apple’, 3]. </span>

```python
def common_elements(list1, list2):
  return list(set(list1) & set(list2))

common_elements([2, 'apple', 10, 'hello', 7], [1, 2, 3, 'apple', 4]) # returnerar ['apple', 2]
```

### <span style="color:turquoise">9. Kim påstår att ”medelvärdet” är ett bättre mått än ”medianvärdet”. Håller du med Kalle? </span>

Det beror på sammanhanget. Medelvärdet påverkas av extrema värden (t.ex om någon tjänar 40 000 miljarder <span style="color:yellow">(shoutout till SD-kvinnan i riksdagen för några år sen)</span>), medan medianen ger ett bättre mått på typisk nivå.
I många fall, t.ex inkomst, är medianen ofta mer rättvisande än medelvärdet. 

Så jag måste tyvärr såga Kims påstående. Median är ofta bättre när datan är snedfördelad. 

Kalle vet jag inte riktigt vad han vill. Antar att det är ett typo och där skulle stå "Håller du med Kim".

```python
inkomster = [22, 23, 24, 25, 26, 27, 28, 29, 100]  # Extremvärde
median = 26
mean = sum(inkomster) / len(inkomster)

plt.figure(figsize=(6, 4))
plt.hist(inkomster, bins=8, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Medel ≈ {mean:.1f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median = {median}')
plt.title("Medelvärde vs Median")
plt.xlabel("Inkomst (tkr)")
plt.ylabel("Antal personer")
plt.legend()
plt.tight_layout()
plt.show()
```

### <span style="color:turquoise">10. Vad används cirkeldiagram för? Ge ett exempel på vad ett företag som Spotify (eller något annat företag som du själv väljer) hade kunnat använda cirkeldiagram för. </span>

Cirkeldiagram används för att visa födelning i procent. 

Exempel: Spotify kan använda ett cirkeldiagram för att visa vilka genrer som är mest populära:

```python
genres = ['Rock', 'Punk', 'Metal', 'Övrigt']
values = [40, 25, 20, 15]

plt.figure(figsize=(6, 6))
plt.pie(values, labels=genres, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#cccccc'])
plt.title("Spotify – Genrefördelning")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### <span style="color:turquoise">11. Vad används linjediagram för? Ge ett exempel på vad ett företag som Spotify (eller något annat företag som du själv väljer) hade kunnat använda linjediagram för.</span>

Linjediagram avnänds för att visa förändingar över tid. 

Exempel: Spotify kan visa hur antalet användare utvecklats månad för månad:

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'Maj']
users = [10, 11, 12.5, 14, 15]

plt.figure(figsize=(6, 4))
plt.plot(months, users, marker='o', linestyle='-', color='blue')
plt.title("Spotify – Aktiva användare över tid")
plt.xlabel("Månad")
plt.ylabel("Miljoner användare")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### <span style="color:turquoise">12. Vad används lådagram för?  </span>

Lådgram `(boxplot)` används för att visa spridning i en datamängd. Det visar t.ex minsta värde, kvartiler, median, och extrema värden.

Exempel: Spotify kan använda boxplots för att analysera lyssningstid per användare och se vilka som lyssnat väldigt lite eller väldigt mycket:

```python
listening_hours = [1.5, 2, 2.5, 3, 3.5, 3.7, 4, 4.2, 4.5, 5, 10, 12]

plt.figure(figsize=(6, 4))
plt.boxplot(listening_hours, vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title("Spotify – Lyssningstid per användare")
plt.xlabel("Timmar per vecka")
plt.tight_layout()
plt.show()
```
